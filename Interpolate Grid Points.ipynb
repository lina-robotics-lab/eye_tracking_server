```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.spatial import ConvexHull
from mpl_toolkits import mplot3d
from region import AcquisitionRegion

%load_ext autoreload
%autoreload 2
```

```python
points = np.array([[1.1,1.0,0.0],[1.0,0.0,1.0],[0.0,0.5,0.0],[0.8,0.4,0.0]])
region = AcquisitionRegion(points)
```

```python
grid = region.grid(0.05)
```

```python
ax = plt.axes(projection='3d')
ax = plt.axes(projection='3d')
for s in region.hull.simplices:
    s = list(s)+[s[0]]
    ax.plot3D(points[s,0],points[s,1],points[s,2])
ax.scatter3D(grid[:,0],grid[:,1],grid[:,2])
plt.show()
```
